1.Sigmoid function

```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x):
  return 1/(1+np.exp(-x))
x=np.linspace(-25,25)
print(x)
plt.plot(x,sigmoid(x))
plt.title("Activation function : Sigmoid")
plt.show()
```

2.Tanh

```python
def tanh(x):
  return np.tanh(x)
x=np.linspace(-10,10,100)
print(x)
plt.plot(x,tanh(x))
plt.title("Activation function : Tanh")
plt.show()
```

3.ReLU

```python
def ReLU(nums):
  return [x if x>=0 else 0 for x in nums]
x=np.linspace(-10,10)
print(x)
plt.plot(x,ReLU(x))
plt.title("Activation function : ReLU")
plt.show()
```

4.Soft-Max

```python
def Soft_max(nums):
  return np.exp(nums)/(np.sum(np.exp(nums),axis=0))
x=np.linspace(-10,10)
print(x)
plt.plot(x,Soft_max(x))
plt.title("Activation function : soft-Max")
plt.show()
```
